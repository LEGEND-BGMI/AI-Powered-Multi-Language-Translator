```python
import torch
import gradio as gr
from transformers import M2M100ForConditionalGeneration, M2M100Tokenizer
from langdetect import detect, DetectorFactory
from gtts import gTTS
import tempfile
import os

DetectorFactory.seed = 0  # stable langdetect results
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)
```

```python
MODEL_NAME = "facebook/m2m100_418M"  # supports 100+ languages
tokenizer = M2M100Tokenizer.from_pretrained(MODEL_NAME)
model = M2M100ForConditionalGeneration.from_pretrained(MODEL_NAME).to(device)
```

```python
LANGUAGES = {
    "auto": "Auto Detect",
    "en": "English",
    "fr": "French",
    "de": "German",
    "es": "Spanish",
    "pt": "Portuguese",
    "it": "Italian",
    "nl": "Dutch",
    "ru": "Russian",
    "zh": "Chinese",
    "ja": "Japanese",
    "ko": "Korean",
    "ar": "Arabic",
    "hi": "Hindi",
    "bn": "Bengali",
    "vi": "Vietnamese",
    "tr": "Turkish",
    "pl": "Polish",
}
```

```python
def translate_text(text, src_lang, tgt_lang):
    if not text.strip():
        return "⚠️ Please enter some text.", None

    # Detect source language if "auto"
    if src_lang == "auto":
        try:
            src_lang = detect(text)
        except Exception:
            return "❌ Could not detect source language.", None

    # Set source language
    tokenizer.src_lang = src_lang

    # Encode input
    encoded = tokenizer(text, return_tensors="pt").to(device)

    # Set target language
    forced_bos = tokenizer.get_lang_id(tgt_lang)

    # Generate translation
    generated = model.generate(
        **encoded,
        forced_bos_token_id=forced_bos,
        max_length=256,
        num_beams=4,
        no_repeat_ngram_size=2,
        early_stopping=True
    )

    translated_text = tokenizer.decode(generated[0], skip_special_tokens=True)

    # Generate speech (TTS) using gTTS
    try:
        tts = gTTS(translated_text, lang=tgt_lang if tgt_lang != "auto" else "en")
        tmp_file = tempfile.NamedTemporaryFile(delete=False, suffix=".mp3")
        tts.save(tmp_file.name)
        audio_path = tmp_file.name
    except Exception as e:
        print("TTS Error:", e)
        audio_path = None

    return translated_text, audio_path
```

```python
with gr.Blocks() as demo:
    gr.Markdown("## 🌍 AI Multi-Language Translator (Text + Speech)")
    gr.Markdown("Type or speak your text → get translation as text + audio 🎤🔊")

    with gr.Row():
        src_lang = gr.Dropdown(choices=list(LANGUAGES.keys()), value="auto", label="Source Language")
        tgt_lang = gr.Dropdown(choices=list(LANGUAGES.keys()), value="en", label="Target Language")

    input_text = gr.Textbox(lines=5, placeholder="Enter or record text...", label="Input Text")
    input_audio = gr.Audio(sources=["microphone"], type="filepath", label="🎤 Record Speech (Optional)")

    output_text = gr.Textbox(lines=5, label="Translation")
    output_audio = gr.Audio(label="🔊 Translation Audio")

    def speech_to_text(audio_path):
        """Placeholder: currently returns None (requires Whisper for STT)."""
        return "⚠️ Speech-to-text requires Whisper model. (Not added yet)" if audio_path else ""

    # Buttons
    btn_text = gr.Button("Translate Text")
    btn_speech = gr.Button("Translate Speech")

    # Events
    btn_text.click(fn=translate_text, inputs=[input_text, src_lang, tgt_lang], outputs=[output_text, output_audio])
    btn_speech.click(fn=lambda a, s, t: translate_text(speech_to_text(a), s, t),
                     inputs=[input_audio, src_lang, tgt_lang], outputs=[output_text, output_audio])

demo.launch(share=True)
```
